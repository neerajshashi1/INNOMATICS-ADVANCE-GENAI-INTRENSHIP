# Problem 1: Unique Words in a Sentence


```python
sentence = "python is easy and python is powerful"

w = sentence.split()
unique_w = set(w)

print("Unique words count:", len(unique_w))
print("Unique words:", unique_w)
```

# Problem 2: Highest Salary from Employee Data


```python
emp = {
    "Rahul": 50000,
    "Anita": 75000,
    "Suresh": 62000
}

highest_sal_employee = max(emp, key=emp.get)

print("Employee with highest salary:", highest_sal_employee,-emp[highest_sal_employee])
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [10, 45, 23, 89, 12, 67]

max = numbers[0]
min = numbers[0]

for num in numbers:
    if num > max:
        max = num
    if num < min:
        min = num

print("Maximum value:", max)
print("Minimum value:", min)
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = [500, 1200, 750, 3000, 1500, 800]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Number of products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "A", "P", "P", "A", "P", "P"]

total_number_days = len(attendance)
total_present_days = attendance.count("P")

percentage = (total_present_days / total_number_days) * 100

print("Attendance Percentage:", percentage, "%")
```

 # Problem 7: Count Character Frequency

```python
text = "python is more powerful"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] +=1
    else: 
        char_count[char] = 1

print("Character frequency: ", char_count)
 
```

# Problem 9: Check if a Key Exists in a Dictionary


```python
student = {
    "name": ['Soniya','sinchana','neeraj'],
    "age": [26,28,22],
    "course": "Data Science"
}

key_to_check = "usn"

if key_to_check in student:
    print("Key exists in dictionary")
else:
    print("Key does not exist")
```
