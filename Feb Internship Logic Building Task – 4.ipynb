# Problem 1: Mobile Recharge Validation System

```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return "Invalid plan selected."
    else:
        return "Recharge Successful!"

# Retry system
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        result = validate_recharge(amount)
        print(result)
        
        if result == "Recharge Successful!":
            break
        else:
            print("Please try again.\n")
            
    except ValueError:
        print("Invalid input. Please enter a number.\n")
```

# Problem 2: Inventory Reorder Alert System

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} → Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} → Stock OK (Stock: {stock})")

products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 8,
    "Soap": 20
}

check_inventory(products)
```

# Problem 3: Student Result Processing System

```python

def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
        
    average = total / len(marks)
    
    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"

marks = [90, 85, 70, 55]
result = process_result(marks)
print(result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python

def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak:
        fare += fare * 0.25  # 25% extra
        
    return fare

# Retry system
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        
        peak = True if peak_input == "yes" else False
        
        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")
        break
        
    except ValueError:
        print("Invalid input. Please try again.\n")
```

# Problem 5: Employee Attendance Eligibility Checker

```python


def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
            
    percentage = (present_days / len(attendance_list)) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P"]
result = check_attendance(attendance)
print(result)
```
